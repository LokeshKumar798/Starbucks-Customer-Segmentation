# 02 — Exploratory Data Analysis (EDA)
**Starbucks Customer Segmentation**

This notebook explores the cleaned dataset to understand distributions, relationships, and patterns in customer demographics and offer behavior.

---

## 1. Import Libraries & Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.figsize'] = (10, 5)

df = pd.read_csv('../data/cleaned_data.csv')
print('Data loaded. Shape:', df.shape)
display(df.head())
```

## 2. Basic Statistics

```python
print('Dataset Info:')
df.info()
print('\nDescriptive Statistics:')
display(df.describe())
```

## 3. Customer Demographics

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Age distribution
axes[0].hist(df['age'].dropna(), bins=30, color='steelblue', edgecolor='white')
axes[0].set_title('Age Distribution')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Count')

# Income distribution
axes[1].hist(df['income'].dropna(), bins=30, color='coral', edgecolor='white')
axes[1].set_title('Income Distribution')
axes[1].set_xlabel('Income')

# Gender distribution
gender_counts = df['gender'].value_counts()
axes[2].bar(gender_counts.index, gender_counts.values, color=['steelblue','coral','mediumseagreen'])
axes[2].set_title('Gender Distribution')
axes[2].set_xlabel('Gender')
axes[2].set_ylabel('Count')

plt.tight_layout()
plt.savefig('../images/demographics.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 4. Offer Event Analysis

```python
# Event type breakdown
event_counts = df['event'].value_counts()
plt.figure(figsize=(8, 4))
sns.barplot(x=event_counts.index, y=event_counts.values, palette='viridis')
plt.title('Distribution of Event Types')
plt.xlabel('Event')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('../images/event_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Offer type breakdown (filter only offer-related events)
offer_df = df[df['offer_type'].notna()]

plt.figure(figsize=(8, 4))
sns.countplot(data=offer_df, x='offer_type', hue='event', palette='Set2')
plt.title('Offer Events by Offer Type')
plt.xlabel('Offer Type')
plt.ylabel('Count')
plt.legend(title='Event')
plt.tight_layout()
plt.savefig('../images/offer_type_events.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 5. Income vs Age by Gender

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df.drop_duplicates('customer_id'), 
                x='age', y='income', hue='gender', alpha=0.5, palette='Set1')
plt.title('Income vs Age by Gender')
plt.xlabel('Age')
plt.ylabel('Income')
plt.tight_layout()
plt.savefig('../images/income_vs_age.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 6. Membership Trend Over Years

```python
membership_trend = df.drop_duplicates('customer_id')['membership_year'].value_counts().sort_index()

plt.figure(figsize=(9, 4))
plt.plot(membership_trend.index, membership_trend.values, marker='o', color='steelblue', linewidth=2)
plt.fill_between(membership_trend.index, membership_trend.values, alpha=0.15, color='steelblue')
plt.title('New Member Registrations per Year')
plt.xlabel('Year')
plt.ylabel('Number of New Members')
plt.tight_layout()
plt.savefig('../images/membership_trend.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 7. Correlation Heatmap

```python
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
corr = df[numeric_cols].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            linewidths=0.5, square=True)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig('../images/correlation_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
```
