## 1. Import Libraries & Load Clustered Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_theme(style='whitegrid')

df = pd.read_csv('../data/clustered_customers.csv', index_col='customer_id')
print('Clustered data loaded. Shape:', df.shape)
print('Clusters:', sorted(df['cluster'].unique()))
```

## 2. Cluster Size Overview

```python
cluster_sizes = df['cluster'].value_counts().sort_index()

plt.figure(figsize=(7, 4))
sns.barplot(x=cluster_sizes.index, y=cluster_sizes.values, palette='tab10')
plt.title('Number of Customers per Cluster')
plt.xlabel('Cluster')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.savefig('../images/cluster_sizes.png', dpi=150, bbox_inches='tight')
plt.show()

print(cluster_sizes)
```

## 3. Cluster Profiles — Mean Feature Values

```python
# Key features to profile
profile_cols = ['age', 'income', 'membership_days', 'offers_received',
                'offers_viewed', 'offers_completed', 'total_transactions',
                'total_spent', 'view_rate', 'completion_rate', 'avg_spend']

cluster_profiles = df.groupby('cluster')[profile_cols].mean().round(2)
display(cluster_profiles)
```

## 4. Visual Cluster Comparison

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
features_to_plot = ['age', 'income', 'total_spent', 'completion_rate', 'avg_spend', 'membership_days']

for ax, feature in zip(axes.flatten(), features_to_plot):
    sns.boxplot(data=df, x='cluster', y=feature, palette='tab10', ax=ax)
    ax.set_title(f'{feature} by Cluster')
    ax.set_xlabel('Cluster')
    ax.set_ylabel(feature)

plt.suptitle('Feature Distribution Across Clusters', fontsize=14, y=1.01)
plt.tight_layout()
plt.savefig('../images/cluster_profiles.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 5. Radar Chart — Cluster Feature Fingerprint

```python
from matplotlib.patches import FancyArrowPatch
import matplotlib.patches as mpatches

# Normalize profiles for radar
radar_features = ['age', 'income', 'total_spent', 'completion_rate', 'avg_spend', 'membership_days']
radar_data = cluster_profiles[radar_features].copy()
radar_norm = (radar_data - radar_data.min()) / (radar_data.max() - radar_data.min())

categories = radar_features
N = len(categories)
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
colors = plt.cm.tab10.colors

for i, (cluster_id, row) in enumerate(radar_norm.iterrows()):
    values = row.tolist()
    values += values[:1]
    ax.plot(angles, values, linewidth=2, label=f'Cluster {cluster_id}', color=colors[i])
    ax.fill(angles, values, alpha=0.1, color=colors[i])

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, size=11)
ax.set_title('Cluster Radar Chart', size=14, pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.tight_layout()
plt.savefig('../images/radar_chart.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 7. Key Takeaways

- **Cluster 0 (High-Value Loyalists):** Older, high-income customers with long membership tenure. They actively complete offers and spend the most. Target with premium, exclusive promotions.

- **Cluster 1 (Deal Seekers):** Mid-income customers who view offers frequently but don't always complete them. Respond well to time-limited discount offers delivered via mobile.

- **Cluster 2 (Passive Customers):** Low engagement and infrequent visits. Need re-activation through personalized outreach and low-barrier informational offers.

- **Cluster 3 (New Members):** Recently joined, moderate spend. Focus on building habits early with welcome offers and loyalty incentives.

---

### Overall Recommendation
Shift from one-size-fits-all promotions to **segment-specific campaigns**. High-value loyalists need exclusivity; deal seekers need urgency; passive customers need re-engagement; new members need habit formation. Personalizing delivery channel (email vs. mobile vs. social) per segment will maximize offer completion rates and ROI.
